```python
x = ['Ada', 'Bays', 'Curie', 'Darden']
y = [17,20,16,21]
plt.bar(x,y, color = "hotpink")
plt.title('Number of students per cohort')
plt.text(0,18, "First Cohort", ha = 'center')
plt.text(1,21,"Second Cohort", ha = 'center')
plt.annotate('Best Cohort', xy=(1,20), xytext=(2,18), arrowprops = {'arrowstyle': '->'})
plt.show()
```

```python
import random
```

```python
x = [random.random() * 100 for _ in range(100)]
x
plt.hist(x, bins = 25, color = 'hotpink')
plt.title("Practicing creating Histogram")
plt.show()
```

```python
x = [random.random() * 100 for _ in range(50)]
x
y1 = [n + random.random() *10 for n in x]
y2 = [100 -n + random.random() * 10 for n in x]
plt.scatter(x, y1, label = "life is good")
plt.scatter(x,y2, label = "is that so?")
plt.legend()
plt.show()
```

```python
time = [1,2,3,4,5]
python_knowledge = [1,1,1,3,4]
sql_knowledge = [0,2,3,3,3]
confidence = [2,3,-1,-4,3]
plt.figure(figsize = (10,10))
plt.plot(time, python_knowledge, label = 'python')
plt.plot(time, sql_knowledge, label = 'sql')
plt.plot(time, confidence, label = "confidence")
plt.xlabel("Time")
plt.legend()
plt.title("My learning experience at codeup")
plt.show()
```
